# نمودار پراکندگی

```python
# (from 0 , to 10 , 20 number)
x=np.random.randint(0 ,10 ,20 )
y=np.random.randint(0 , 10 ,20)
plt.scatter(x,y ,color='green')
```

```python
x=np.random.randint(0 ,10 ,20 )
y=np.random.randint(0 , 10 ,20)
plt.scatter(x,y ,marker="^")
```

```python
x=[1,2,3,5,8,4]
y=[10,4,7,5,3,2]
label=[2,2,2 , 1,1,1]     # change the color 
plt.scatter(x, y, marker='^', c=label)
```

```python
activities = ['eat', 'sleep', 'work', 'play'] 
slices = [2,8,10,20] 
colors = ['r', 'y', 'g', 'b'] 
plt.pie(slices, labels = activities, colors=colors)
plt.legend() 
```

```python
x=np.linspace(-2*np.pi,2*np.pi ,100)  
plt.subplot(2,2,1)
y1=x
plt.plot(x,y1)
plt.subplot(2,2,2)
y2=np.sin(x)
plt.plot(x,y2)
plt.subplot(223)
y3=np.cos(x)
plt.plot(x,y3)
plt.subplot(224)
y4=x**2
plt.plot(x,y4)
```
